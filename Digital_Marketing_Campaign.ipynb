```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
###### 2 Logistic Regression
x = data.drop("Conversion", axis=1)
y = data["Conversion"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=5)

numeric_column = x_train.select_dtypes(np.number).columns
categorical_column = x_train.select_dtypes("object").columns
scaler = StandardScaler()
encoder = OneHotEncoder(handle_unknown="ignore", drop="first")

transform = ColumnTransformer([
    ("num", scaler, numeric_column),
    ("cat", encoder, categorical_column)
])
x_train_transform = transform.fit_transform(x_train)
x_test_transform  = transform.transform(x_test)
```

```python
#check the multicollinearity assumption with the VIF value must be <10
def calc_vif(x, feature_names):
    vif = pd.DataFrame()
    vif['Feature'] = feature_names
    vif['VIF'] = [variance_inflation_factor(x, i) for i in range(x.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by='VIF', ascending=False)
    return vif

calc_vif(x_train_transform, feature_names)
```

No variable has a value greater than 10, so there is no need to delete any variables. This data also meets the multicollinearity assumption for logistic regression analysis.

```python
#Building model
lr = LogisticRegression()
lr.fit(x_train_transform, y_train)
```

```python
#Evaluation
#Model Accuracy
prediction = lr.predict(x_test_transform)
accuracy = accuracy_score(y_test, prediction)
print("Accuracy:", accuracy)
```

The accuracy of the model created is 89.54% which seems to be decent.

```python
#Mean Accuracy
scores = cross_val_score(lr, x_train_transform, y_train, cv=10)
mean_accuracy = scores.mean()

print("Mean Accuracy:", mean_accuracy)
```

The mean accuracy of the model created is correctly predict outcome (conversion or not) for 88,59% which shows strong performance.

```python
#Classification Report
print(classification_report(y_test, prediction))
```

- The model performs very well at predicting class 1 (the positive class), with high precision (91%) and recall (98%), leading to a strong F1-score (94%).
- However, it struggles significantly with predicting class 0 (the negative class), as indicated by low precision (63%) and very low recall (26%). This suggests that many actual negatives are being misclassified as positives.
